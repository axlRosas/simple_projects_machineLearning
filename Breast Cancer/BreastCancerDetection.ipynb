## Splitting the dataset

```python
X=dataset.iloc[:,1:-1].values
```

```python
X.shape
```

```python
#target variable
y=dataset.iloc[:,-1].values
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

# Paso 2: Creacion del modelo

 modelos lineales: https://scikit-learn.org/stable/modules/classes.html#module-sklearn.linear_model

```python
from sklearn.linear_model import LogisticRegression
```

```python
model_lr=LogisticRegression(random_state=0)
```

```python
model_lr.fit(X_train,y_train)
```

```python
y_pred=model_lr.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,f1_score,precision_score, recall_score
```

```python
acc=accuracy_score(y_test,y_pred)
f1=f1_score(y_test,y_pred)
prec=precision_score(y_test,y_pred)
rec=recall_score(y_test,y_pred)
```

```python
results=pd.DataFrame([["Logistic Regression",acc,f1,prec,rec]],
                    columns=["Model","Accuracy","F1_Score","Precision","Recall"])
```

```python
results
```

```python
cm=confusion_matrix(y_test,y_pred)
print(cm)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
accuracies=cross_val_score(estimator=model_lr,X=X_train,y=y_train,cv=10)
```

```python
accuracies
```

```python
print("la precision es {:.2f}%".format(accuracies.mean()*100))
print("La desviacion estandar es {:.2f}%".format(accuracies.std()*100))
```

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model_rf=RandomForestClassifier(random_state=0)
model_rf.fit(X_train,y_train)
```

```python
y_pred=model_rf.predict(X_test)
```

```python
acc=accuracy_score(y_test,y_pred)
f1=f1_score(y_test,y_pred)
prec=precision_score(y_test,y_pred)
rec=recall_score(y_test,y_pred)
```

```python
results_RF=pd.DataFrame([["Random_Forest",acc,f1,prec,rec]],
                    columns=["Model","Accuracy","F1_Score","Precision","Recall"])
```

```python
results_RF
```

```python
results=results.append(results_RF,ignore_index=True)
```

```python
results
```

```python
cm=confusion_matrix(y_test,y_pred)
print(cm)
```

### Cross validation

```python
accuracies=cross_val_score(estimator=model_rf,X=X_train,y=y_train,cv=10)
```

```python
accuracies
```

```python
print("la precision es {:.2f}%".format(accuracies.mean()*100))
print("La desviacion estandar es {:.2f}%".format(accuracies.std()*100))
```

# Part 3: Fine tunning Logistic Regression

```python
from sklearn.model_selection import RandomizedSearchCV
#https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.RandomizedSearchCV.html#sklearn.model_selection.RandomizedSearchCV
```

```python
parameters={
    'penalty': ['l1','l2','elasticnet','none'],
    'C':[0.25,0.5,0.75,1.0,1.25,1.5,1.75,2.0],
    'solver': ['lbfgs','liblinear', 'newton-cg', 'newton-cholesky', 'sag','saga']
}
```

```python
random_search=RandomizedSearchCV(estimator=model_lr,param_distributions=parameters,
                                 n_iter=10,scoring='roc_auc',n_jobs=-1, cv=10,verbose=3)
```

```python
random_search.fit(X_train,y_train)
```

```python
random_search.best_estimator_
```

```python
random_search.best_score_
```

```python
random_search.best_params_
```

# Part 5 Final model (Logistic Regression)

```python
modelo=LogisticRegression(C=1.25, penalty='l1', random_state=0, solver='saga')
```

```python
modelo.fit(X_train,y_train)
```

```python
y_pred=modelo.predict(X_test)
```

```python
acc=accuracy_score(y_test,y_pred)
f1=f1_score(y_test,y_pred)
prec=precision_score(y_test,y_pred)
rec=recall_score(y_test,y_pred)
```

```python
results_BLR=pd.DataFrame([["Best LR",acc,f1,prec,rec]],
                    columns=["Model","Accuracy","F1_Score","Precision","Recall"])
```

```python
results=results.append(results_BLR,ignore_index=True)
```

```python
results
```
